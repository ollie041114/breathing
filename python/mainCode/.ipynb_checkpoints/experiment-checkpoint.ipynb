```python
from neulog import Device

# from gzp import save
import time
from breathingRate import breathing_rate_estimation

"""
Sample GSR experiment.

Gathers data over two phases. Use a keyboard interrupt (control-c) to end a phase.

Saves data to disk afterwards.
"""
```

```python
# Create a new device
device = Device(port = "COM5")
```

```python
data = []

def experimentCode():
    global data 
    global device 

    while True:
        v = device.getSensorsData('Respiration', 1)
        if v == False:
            next
        if v != False: 
            variable = float(str(v))
            data.append(variable)
            print(v, end='\r', flush=True)
            break


samplingPeriod = 0.1 # sample every 0.5 seconds
device.setSampleRate(samplingPeriod)
start = time.time()
now = start 
while True:
    if (time.time() - now) >= samplingPeriod:
        now = time.time()
        experimentCode()
    if time.time() - start >= 60:
        break   


### save to data.csv as df 
import pandas as pd
df = pd.DataFrame(data)


df.to_csv('data/data3.csv')
```

```python
import numpy as np
from scipy.fft import rfft, rfftfreq


sample_rate = 1/0.1


def calculate_biofeedback_score(power_spectrum, target_range):
    # Extract the frequencies and power values from the power spectrum
    frequencies = power_spectrum[:, 0]
    power = power_spectrum[:, 1]

    # Calculate the area under the curve (AUC) of the power spectrum within the
    # target range
    auc_target = np.trapz(power[(frequencies >= target_range[0]) & (frequencies <= target_range[1])], frequencies[(
        frequencies >= target_range[0]) & (frequencies <= target_range[1])])

    # Calculate the AUC of the power spectrum outside of the target range
    auc_outside = np.trapz(power[(frequencies < target_range[0]) | (frequencies > target_range[1])], frequencies[(
        frequencies < target_range[0]) | (frequencies > target_range[1])])

    # Calculate the biofeedback score as the ratio of the AUC within the target range
    # to the AUC outside of the target range
    biofeedback_score = auc_target / auc_outside

    return biofeedback_score


def calculate_power_spectrum(signal):
    # Calculate the FFT of the signal
    N = len(signal)
    yf = rfft(signal)
    xf = rfftfreq(N, 1/sample_rate)

    yf[0]= 0
    xf[0] = 0

    return np.array([xf, np.abs(yf) ** 2]).T


# Define the target range (breaths per minute)
target_range = (4, 12)

# Convert the target range to breaths per second
target_range_breaths_per_second = [bpm / 60 for bpm in target_range]

print(target_range_breaths_per_second)
# Calculate the power spectrum of the breathing signal
power_spectrum = calculate_power_spectrum(data)

# Calculate the biofeedback score from the power spectrum and target range
biofeedback_score = calculate_biofeedback_score(
    power_spectrum, target_range_breaths_per_second)

print(biofeedback_score)
```

```python
%matplotlib notebook

import numpy as np
import matplotlib
from matplotlib import pyplot as plt

# Calculate the power spectrum of the window
power_spectrum = calculate_power_spectrum(data)

xf = power_spectrum[:, 0]
yf = np.abs(power_spectrum[:, 1])


# You probably won't need this if you're embedding things in a tkinter plot...
# plt.ion()

fig = plt.figure()
ax = fig.add_subplot(111)

line1, = ax.plot(xf, yf, 'r-') # Returns a tuple of line objects, thus the comma
plt.axvspan(target_range_breaths_per_second[0], target_range_breaths_per_second[1], alpha=0.1, color='red')
# for phase in np.linspace(0, 10*np.pi, 500):
#     line1.set_ydata(np.sin(x + phase))
#     fig.canvas.draw()
#     fig.canvas.flush_events()


def process_breathing_data(data, window_size, hop_size):
    # Convert the window size and hop size from seconds to samples
    window_size_samples = int(window_size * sample_rate)
    hop_size_samples = int(hop_size * sample_rate)

    # Initialize a list to store the biofeedback scores
    biofeedback_scores = []

    # Iterate over the data using a window-sliding method
    for i in range(0, len(data) - window_size_samples, hop_size_samples):
        # Extract the current window of data
        window = data[i:i+window_size_samples]

        # Calculate the power spectrum of the window
        power_spectrum = calculate_power_spectrum(window)
        
        xf = power_spectrum[:, 0]
        yf = power_spectrum[:, 1]
        
        line1.set_xdata(xf)
        line1.set_ydata(np.abs(yf))
        fig.canvas.draw()
        fig.canvas.flush_events()

        # Find the dominant frequency
        dominant_freq = np.argmax(np.abs(yf))
        # print("Dominant frequency is", xf[dominant_freq], "Hz")

        # Calculate the biofeedback score from the power spectrum and target range
        biofeedback_score = calculate_biofeedback_score(power_spectrum, target_range_breaths_per_second)

        # Append the biofeedback score to the list
        biofeedback_scores.append(biofeedback_score)

    return biofeedback_scores


biofeedback_scores = process_breathing_data(data, window_size=30, hop_size=2)
```
